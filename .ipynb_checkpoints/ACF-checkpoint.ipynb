# ACF from Scratch

```python
import numpy as np
import pandas as pd
from statsmodels.tsa.stattools import acf
import matplotlib.pyplot as plt
plt.style.use("seaborn-colorblind")

def autocorr_by_hand(x, lag):
    # Slice the relevant subseries based on the lag
    y1 = x[:(len(x)-lag)]
    y2 = x[lag:]
    # Subtract the subseries means
    sum_product = np.sum((y1-np.mean(y1))*(y2-np.mean(y2)))
    # Normalize with the subseries stds
    return sum_product / ((len(x) - lag) * np.std(y1) * np.std(y2))

def acf_by_hand(x, lag):
    # Slice the relevant subseries based on the lag
    y1 = x[:(len(x)-lag)]
    y2 = x[lag:]
    # Subtract the mean of the whole series x to calculate Cov
    sum_product = np.sum((y1-np.mean(x))*(y2-np.mean(x)))
    # Normalize with var of whole series
    return sum_product / ((len(x) - lag) * np.var(x))

x = np.linspace(0,100,101)

results = {}
nlags=20
results["acf_by_hand"] = [acf_by_hand(x, lag) for lag in range(nlags)]
results["autocorr_by_hand"] = [autocorr_by_hand(x, lag) for lag in range(nlags)]
results["autocorr"] = [pd.Series(x).autocorr(lag) for lag in range(nlags)]
results["acf"] = acf(x, unbiased=True, nlags=nlags-1)

pd.DataFrame(results).plot(kind="bar", figsize=(10,5), grid=True)
plt.xlabel("lag")
plt.ylim([-1.2, 1.2])
plt.ylabel("value")
plt.show()
```

# load and prepare Data

```python
cc = pd.read_csv("Data/data_clean2.csv")
len(cc.cycle_index.unique())

#missing days interpolated
md = pd.read_csv("Data/missingdays_interpolated.csv")
len(md.cycle_index.unique())
```

```python
def cycle_transform(df, value, save = True):

    value_ = df[["cycle_day", value, "cycle_index"]]

    value_zero = value_.pivot_table(index = "cycle_day", columns="cycle_index", values = value_).fillna(0)[value].reset_index()
    value_nan = value_.pivot_table(index = "cycle_day", columns="cycle_index", values = value_)[value].reset_index()

    if save == True:
        value_zero.to_csv("Data/"+value+"_zero.csv")
        value_nan.to_csv("Data/"+value+"_nan.csv")
    return value_zero, value_nan
```

```python
hr29_norm_zero, hr29_norm_nan = cycle_transform(md[md.cycle_length == 29], "hr_norm", True)
hr29_norm_zero2, hr29_norm_nan2 = cycle_transform(cc[cc.cycle_length == 29], "hr_norm", True)

br29_norm_zero, br29_norm_nan = cycle_transform(md[md.cycle_length == 29], "br_norm", True)
br29_norm_zero2, br29_norm_nan2 = cycle_transform(cc[cc.cycle_length == 29], "br_norm", True)

st29_norm_zero, st29_norm_nan = cycle_transform(md[md.cycle_length == 29], "st_norm", True)
st29_norm_zero2, st29_norm_nan2 = cycle_transform(cc[cc.cycle_length == 29], "st_norm", True)
```

# ACF and PACF for some cycles

```python
def plot_acf(timeSeries, column):
    from statsmodels.tsa.stattools import acf, pacf
    plt.style.use('bmh')
    results = {}
    nlags=len(timeSeries)


    for cycle in timeSeries.columns[[column]]:
        print("cycle_index: ",cycle)
        timeSeries[cycle].plot( figsize=(15,5))
        plt.xlabel("day")
        plt.ylabel("value")
        plt.xticks(range(0,29))


        results["acf"] = acf(timeSeries[cycle], unbiased=True, nlags=nlags-1)
      #  results["pacf"] = pacf(timeSeries[cycle], nlags=nlags-1, method='ols')

        pd.DataFrame(results).plot(kind="bar", figsize=(15,5), grid=True)
        plt.xlabel("lag")
        plt.ylim()
        plt.ylabel("value")
        plt.show()
```

```python
def plot_acf_all(hr,br,st, column):
    from statsmodels.tsa.stattools import acf, pacf
    plt.style.use('bmh')
    results = {}
    nlags=len(hr)


    for cycle in hr.columns[[column]]:
        print("cycle_index: ",cycle)
        hr[cycle].plot( figsize=(15,5),label = "hr")
        br[cycle].plot( figsize=(15,5),label = "br")
        st[cycle].plot( figsize=(15,5),label = "st")
        plt.xlabel("day")
        plt.ylabel("value normalized")
        plt.legend()
        plt.xticks(range(0,29))


        results["acf_hr"] = acf(hr[cycle], unbiased=True, nlags=nlags-1)
        results["acf_br"] = acf(br[cycle], unbiased=True, nlags=nlags-1)
        results["acf_st"] = acf(st[cycle], unbiased=True, nlags=nlags-1)
        
        
        #Plot ACF: 
        pd.DataFrame(results).plot(kind = "bar", figsize=(15,5))


        
        #results["pacf"] = pacf(timeSeries[cycle], nlags=nlags-1)
        plt.title('Autocorrelation Function')
        plt.axhline(y=0,linestyle='--',color='gray')
        plt.axhline(y=-1.96/np.sqrt(len(results["acf_hr"])),linestyle='--',color='gray')
        plt.axhline(y=1.96/np.sqrt(len(results["acf_hr"])),linestyle='--',color='gray')
        plt.xlabel("lag")
        plt.ylim()
        plt.ylabel("value")
        plt.show()
    return results
```

```python
br = br29_norm_zero.drop("cycle_day", axis = 1)
hr = hr29_norm_zero.drop("cycle_day", axis = 1)
st = st29_norm_zero.drop("cycle_day", axis = 1)
```

```python
acf = plot_acf_all(br,hr,st, 10)
```

```python
import numpy as np
import statsmodels.api as sm
lowess = sm.nonparametric.lowess
x = np.random.uniform(low = -2*np.pi, high = 2*np.pi, size=500)
y = np.sin(x) + np.random.normal(size=len(x))
z = lowess(x,y)
w = lowess(x,y, frac=1./3)
```
