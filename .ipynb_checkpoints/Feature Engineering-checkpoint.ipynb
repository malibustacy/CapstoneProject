```python
df = pd.read_csv("Data/data_clean.csv").drop("Unnamed: 0", axis =1)
#df[df["br_norm"].isnull()]
```

```python
features = df[["user", "cycle_index","cycle_day", "skin_temperature", "hr", "br", "cycle_number"]]
```

```python
### perform extraction, imputing and filtering at the same time
from tsfresh import extract_features

features_ = extract_features(features, column_id='cycle_index', column_sort='cycle_day')
```

```python
features_.isnull()
```

```python
from tsfresh import select_features
from tsfresh.utilities.dataframe_functions import impute

impute(features_)
features_filtered = select_features(features_)
```
