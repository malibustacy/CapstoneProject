```python
# %load cycle_transform.py
def cycle_transform(df, value, save = True):

    value_ = df[["cycle_day", value, "cycle_index"]]
    #print(sorted(value_.cycle_index.unique()))

    value_zero = value_.pivot_table(index = "cycle_day", columns="cycle_index", values = value_).fillna(0)
    #print("value_zero:",value_zero)
    value_nan = value_.pivot_table(index = "cycle_day", columns="cycle_index", values = value_)

    if save == True:
        value_zero.to_csv("Data/"+value+"_zero.csv")
        value_nan.to_csv("Data/"+value+"_nan.csv")
    return value_zero, value_nan
```

```python
def cluster_percentage_boxplot(n_clusters, user_cycle_df, clusters_hierarchical):
    mergedclusters = pd.DataFrame()
    
    user_cycle_df["cluster_name"] = clusters_hierarchical
    clusters = user_cycle_df[["user", "cluster_name"]]


    clusters = user_cycle_df.groupby(["user", "cluster_name"]).count().reset_index().drop(["cycle_number", "cycle_length"], axis = 1)
    #print("clusters", clusters)
    clusters.columns = ["user", "cluster_name", "count"]
    user_counts = user_cycle_df.groupby(["user"]).count()["cycle_length"]
    user_counts = user_counts#[user_counts.values > 2]
    #print("user_counts", len(user_counts))
    # loop through the clusters
    for cluster in range(1,n_clusters+1):
        #print(cluster)
        #print(clusters[clusters["#cluster"] == cluster])
        
        #merge clusters with current cluster number (cluster) with user_count
        mergedcluster1 = pd.merge(clusters[clusters["cluster_name"] == cluster], pd.DataFrame(user_counts.reset_index()), on="user")
        #give a right name
        mergedcluster1["count_total"] = mergedcluster1["cycle_length"]
        #calculate the percentage by dividing counted users through the total amount of cycles
        mergedcluster1["percentage"] =  100 * mergedcluster1["count"]/mergedcluster1["count_total"]
        #add a column where the cluster number for the percentage is written
        #mergedcluster1["cluster"] = cluster
        #add the percentage values of this cluster to the main dataframe
        mergedclusters = pd.concat([mergedclusters, mergedcluster1], ignore_index=True)

    with plt.style.context(('bmh')):
        fig = plt.figure(figsize=(15,10))
        plt.style.use(('bmh'))
        sns.boxplot(x="cluster_name", y="percentage", data=mergedclusters)

    return mergedclusters.drop("cycle_length", axis = 1), user_cycle_df
```

```python
def DTWDistance(s1, s2):
    DTW={}
    w = 2
    w = max(w, abs(len(s1)-len(s2)))

    for i in range(-1,len(s1)):
        for j in range(-1,len(s2)):
            DTW[(i, j)] = float('inf')
    DTW[(-1, -1)] = 0

    for i in range(len(s1)):
        for j in range(max(0, i-w), min(len(s2), i+w)):
            dist= (s1[i]-s2[j])**2
            DTW[(i, j)] = dist + min(DTW[(i-1, j)],DTW[(i, j-1)], DTW[(i-1, j-1)])

    return np.sqrt(DTW[len(s1)-1, len(s2)-1])
```

```python
def print_heatmap2(timeSeries, user_cycle_df, cycle_length, metric, method, first_cycle = True):

    import seaborn as sns; sns.set(color_codes=True)

    if cycle_length == "all":
        users = user_cycle_df[["user","cycle_index","cycle_number"]]
        users = users.set_index(users.cycle_index)
        figsize = (15,15)

    else:
        users = user_cycle_df[user_cycle_df.cycle_length == cycle_length][["user","cycle_index","cycle_number"]]
        users = users.set_index(users.cycle_index)
        figsize = (15,15)

    lut = dict(zip(users.user.unique(), sns.husl_palette(len(set(users.user)),l=0.5)))
    row_colors = users.user.map(lut).rename("user")
    #print("row_colors", row_colors)
    
    if first_cycle == True:
        #first cycle colors
        max_cycle_numbers = len(users.cycle_number.unique())-7
        colors = "ryyyggg"+("w"*max_cycle_numbers)
        lut2 = dict(zip(sorted(users.cycle_number.unique()), colors))
        row_colors2 = users.cycle_number.map(lut2).rename("first cycle")
        #print("row_colors2", row_colors2)
    
        row_color = [row_colors2, row_colors]
        
    if first_cycle == False:
        row_color = row_colors

    mask = timeSeries.isnull()
    with sns.axes_style("white"):
        ax = sns.clustermap(timeSeries.fillna(0),
                    metric=metric,method=method,
                    col_cluster = False,
                    figsize=figsize,
                    row_colors = row_color,
                    mask = mask,
                    )
        plt.setp(ax.ax_heatmap.get_yticklabels(), rotation=0)  # For y axis
        plt.setp(ax.ax_heatmap.get_xticklabels(), rotation=90) # For x axis
        plt.title('', fontsize=12)
        plt.suptitle(method+"-"+metric+" clustered cylces ("+str(cycle_length)+" Days)", fontsize=20)
        plt.legend(bbox_to_anchor=(1, 1), loc=2)
        #plt.savefig('heatmap2.png')
```
