```python
import pandas as pd
import numpy as np
import scipy
import os
import os.path
from matplotlib import cm
from IPython.display import display, HTML

import matplotlib
%matplotlib inline
%config InlineBackend.figure_format = 'retina' 
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.dates as mdates
import calendar
```

```python
#check missing values
def num_missing(x):
    return sum(x.isnull())
```

```python
#df_regLen = df2.loc[(df2.cycle_length >= 20)&(df2.cycle_length <= 36)]
#df_regLen.to_csv("Data/regLen.csv")

#load cycle data with regular length
df_regLen = pd.read_csv("Data/regLen.csv").drop(["Unnamed: 0", "index"], axis = 1)
#print(df_regLen.shape)
#print(df_regLen.head())

print("Missing values per column:")
print(df_regLen.apply(num_missing, axis=0))

#group cycles by user and cycle number
groupby_cycles = df_regLen.groupby(["user","cycle_number"]).describe()
groupby_cycles.head()
```

```python
#reset index to change user and cycle number into normal columns
groupby_cycles = pd.DataFrame(groupby_cycles[["cycle_day", "cycle_length"]]).reset_index()
```

```python
#get indeces of incomplete cycles
index_incompleteCycles = groupby_cycles[groupby_cycles.cycle_day["count"] != groupby_cycles.cycle_length["mean"]].index

#save incomplete cycles with regular length (20 - 36 days) in new dataframe
missingdays_df = groupby_cycles.iloc[index_incompleteCycles][["user", "cycle_number"]]
```

```python
# group missingdays_df by cycles
missingdays = pd.DataFrame()
for user in np.unique(missingdays_df.user.values):
    for cycle in missingdays_df[missingdays_df.user == user].cycle_number.values:
        #print("user: ", user)
        #print("cycle: ", cycle)
        missingdays = missingdays.append(df_regLen[(df_regLen["user"] == user)&(df_regLen["cycle_number"] == cycle)])
        #print(gold[["user", "cycle_number", "date"]])
```

```python
# look at cycle length description to compare "count" (number of days) and "mean" (actual cycle length)
missingdays.groupby("cycle_index").describe()["cycle_length"].head()
```

```python
#calculate "completeness"
#completeness = cycle_length-count/ cycle_length-mean

completeness = missingdays.groupby("cycle_index").describe()["cycle_length"]["count"]/missingdays.groupby("cycle_index").describe()["cycle_length"]["mean"]
```

```python
# reset index and create dataframe from groupby object
completeness = completeness.reset_index()
completeness = pd.DataFrame(completeness)
```

```python
completeness = completeness.rename(columns={0 : "completeness"})
```

```python
#indeces of cycles which have less or equal 20% of data missing
cycle_indeces_usable = completeness[completeness.completeness >= 0.8]["cycle_index"]
```

```python
print("{} cycles with at least 80% of cycle days.".format(cycle_indeces_usable.size))
```

```python
missingdays_usable = missingdays[missingdays['cycle_index'].isin(cycle_indeces_usable)]
```
